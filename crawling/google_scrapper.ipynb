```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time

driver = webdriver.Chrome('/usr/local/Cellar/chromedriver/chromedriver')
driver.get('https://www.visitjeju.net/kr/detail/list?menuId=DOM_000001718001000000&cate1cd=cate0000000002#p1&region2cd&pageSize=6&sortListType=reviewcnt&viewType=thumb')

for i in range(1,20):
    place_id = "//*[@id='content']/div/div[2]/div[5]/ul/li["+str(i)+"]/dl/dt/a"
    print(place_id)
    place_tag = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, place_id)))
    place_tag.click()
    
    name_id = "//*[@id='content']/div/div[2]/div[1]/div[2]/div[1]/h3"
    name_tag = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, name_id)))
    print(name_tag.text)

    address_id = "//*[@id='content']/div/div[2]/div[1]/div[2]/div[5]/div[1]/p[2]"
    address_tag = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, address_id)))

    print(address_tag.text)
    
    telephone_id = "//*[@id='content']/div/div[2]/div[1]/div[2]/div[5]/div[2]/p[2]"
    telephone_tag = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, telephone_id)))
    print(telephone_tag.text)

    detail_tag = driver.find_element_by_css_selector("#detailTap > button.tab_btn.tab_btn2")
    detail_tag.click()
    
    
    
    place_time_id = "//*[@id='content']/div/div[3]/div[2]/table/tbody/tr[3]/td"
    
    try:
        place_time_tag = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, place_time_id)))
        print(place_time_tag.text)
    except NoSuchElementException:
        print("No element Found")
    
    print("=====================")

    driver.back()
    driver.back()


driver.quit()
```
